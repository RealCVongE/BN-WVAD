```python
def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    return quick_sort(left) + middle + quick_sort(right)

names = ["Charlie", "Alice", "Eve", "Bob"]
sorted_names = quick_sort(names)
print(sorted_names)
```

```python
with open("sorted_normal_train.txt", "w") as file:
    # 리스트의 각 항목을 줄 바꿈과 함께 파일에 쓰기
    for item in sorted_names:
        file.write(f"{item}\n")
```

```python
import numpy as np

# 파일 경로를 입력하세요
file_path = 'sort_normal.list'

# 경로 목록을 저장할 리스트
path_list = []

# 차원 길이가 다른 경로를 저장할 리스트
different_shape_paths = []

# 텍스트 파일에서 경로 읽기
with open(file_path, 'r') as file:
    path_list = [line.strip() for line in file.readlines()]

# 기준이 될 배열의 shape 길이 (첫 번째 로드된 배열의 shape 길이로 초기화)
reference_shape_length = None

# 각 파일에 대해
for path in path_list:
    try:
        # NumPy 배열 로드
        array = np.load(path)

        # 참조 shape 길이가 설정되지 않았으면 현재 배열의 shape 길이를 사용
        if reference_shape_length is None:
            reference_shape_length = len(array.shape)
        # 현재 배열의 shape 길이가 참조와 다르면 목록에 추가
        if len(array.shape) != reference_shape_length:
            print(path)
            different_shape_paths.append(path)
    except Exception as e:
        print(f"Error loading {path}: {e}")

# 차원 길이가 다른 경로 출력
print("Paths with different shape lengths:", different_shape_paths)
```
