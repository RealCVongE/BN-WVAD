```python
def extract_path(file, path):
    for filename in os.listdir(path):
        file_path = os.path.join(path, filename)  # 파일의 전체 경로
        if os.path.isfile(file_path):  # 파일이 맞는지 확인
            file.append(file_path)

    return file
```
